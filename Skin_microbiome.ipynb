```r
# Aggregate to phylum level
physeq_phylum <- tax_glom(physeq, taxrank = "Phylum")

# Get OTU table as matrix
phyla_matrix <- as(otu_table(physeq_phylum), "matrix")

# Ensure taxa are rows
if (!taxa_are_rows(physeq_phylum)) {
  phyla_matrix <- t(phyla_matrix)
}

# Remove phyla with zero variance
phyla_matrix <- phyla_matrix[apply(phyla_matrix, 1, sd) != 0, ]

# Correlation matrix
cor_matrix <- cor(t(phyla_matrix), method = "pearson")

# Heatmap of correlations
pheatmap(cor_matrix, main = "correlation matrix of ASVs")
```

```r
#library(phyloseq)
library(ggplot2)

# 1. Aggregate at phylum level
ps_phylum <- tax_glom(physeq, taxrank = "Phylum")

# 2. Convert OTU table to data frame
otu_phylum <- as.data.frame(otu_table(ps_phylum))

# 3. Calculate total abundance per phylum
tax_phylum <- as.data.frame(tax_table(ps_phylum))
otu_phylum$Phylum <- tax_phylum$Phylum
otu_phylum$TotalAbundance <- rowSums(otu_phylum[, sapply(otu_phylum, is.numeric)])

# 4. Create phyla_df for ggplot
phyla_df <- aggregate(TotalAbundance ~ Phylum, data = otu_phylum, sum)
colnames(phyla_df) <- c("Phylum", "Abundance")

# 5. Pie Chart
phyla_df$Percentage <- round(100 * phyla_df$Abundance / sum(phyla_df$Abundance), 1)
phyla_df$Label <- paste0(phyla_df$Phylum, " (", phyla_df$Percentage, "%)")

ggplot(phyla_df, aes(x = "", y = Abundance, fill = Label)) +
  geom_bar(width = 1, stat = "identity") +
  coord_polar("y") +
  labs(title = "Microbial Community Composition") +
  theme_void() +
  theme(legend.title = element_blank())

```

```r
library(phyloseq)
library(pheatmap)
library(RColorBrewer)

# Aggregate to phylum level
ps_phylum <- tax_glom(ps, taxrank = "Phylum")
otu_phylum <- as.data.frame(otu_table(ps_phylum))
tax_phylum <- as.data.frame(tax_table(ps_phylum))

# Transpose: rows = samples, columns = phyla
otu_phylum_t <- t(otu_phylum)

# Remove columns with zero variance (to avoid correlation errors)
otu_phylum_t <- otu_phylum_t[, apply(otu_phylum_t, 2, function(x) sd(x) != 0)]

# Compute correlation matrix
cor_matrix <- cor(otu_phylum_t, method = "pearson", use = "pairwise.complete.obs")

# Create phylum names (remove duplicates if any)
tax_phylum_filtered <- tax_phylum[colnames(otu_phylum_t), , drop=FALSE]
phyla <- tax_phylum_filtered$Phylum
names(phyla) <- colnames(otu_phylum_t)

# Annotation for heatmap
annotation <- data.frame(Phylum = phyla)
rownames(annotation) <- names(phyla)

# Use colorRampPalette to make as many colors as needed
phyla_unique <- unique(phyla)
phyla_colors <- setNames(colorRampPalette(brewer.pal(9, "Set1"))(length(phyla_unique)), phyla_unique)
ann_colors <- list(Phylum = phyla_colors)

# Plot heatmap
pheatmap(cor_matrix,
         main = "Phylum Correlation Heatmap",
         color = colorRampPalette(c("blue", "white", "red"))(100),
         annotation_row = annotation,
         annotation_col = annotation,
         annotation_colors = ann_colors,
         border_color = NA)
```

```r
#Dendrogram of Phyla
# Aggregate to phylum level
ps_phylum <- tax_glom(ps, taxrank = "Phylum")

# Create Bray-Curtis distance matrix
dist_bc <- phyloseq::distance(ps_phylum, method = "bray")

# Hierarchical clustering
clust <- hclust(as.dist(dist_bc), method = "average")

# Plot dendrogram
plot(clust, main = "Dendrogram of Phylum-Level Microbial Communities", xlab = "", sub = "", cex = 0.8)
```

```r
# clade coloring by phylum
# Load required libraries
library(phyloseq)
library(dendextend)

# Aggregate to phylum level
ps_phylum <- tax_glom(ps, taxrank = "Phylum")

# Extract the taxonomy
tax_data <- as.data.frame(tax_table(ps_phylum))
otu_names <- taxa_names(ps_phylum)

# Ensure phylum names are associated with tips
phylum_labels <- tax_data[otu_names, "Phylum"]

# Create Bray-Curtis distance matrix
dist_bc <- phyloseq::distance(ps_phylum, method = "bray")

# Hierarchical clustering
clust <- hclust(as.dist(dist_bc), method = "average")

# Convert to dendrogram object
dend <- as.dendrogram(clust)

# Label tips by Phylum
labels_colors(dend) <- as.numeric(as.factor(phylum_labels))

# Color branches by Phylum
dend <- dend %>%
  set("labels_colors", as.numeric(as.factor(phylum_labels))) %>%
  set("branches_k_color", k = length(unique(phylum_labels))) # number of phyla

# Plot
plot(dend,
     main = "Dendrogram of Phylum-Level Microbial Communities",
     ylab = "Bray-Curtis Distance")

legend("topright", legend = unique(phylum_labels),
       fill = 1:length(unique(phylum_labels)), cex = 0.4)
```
